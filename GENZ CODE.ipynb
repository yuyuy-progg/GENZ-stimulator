```python
# 1. user data
user = {}
user["name"] = input("ğŸ‘¤ Enter your name: ")
user["age"] = input(" Enter your age: ")

# 2. bot responses
responses = {
    "hi": "Hey there! ğŸ˜Š",
    "hello": "Hi! How are you doing? ğŸ‘‹",
    "how are you?": "I'm doing fine, thanks! You?",
    "what is your name": "I'm your GenZ bot ğŸ˜",
    "i am good": "great to know ğŸ’¬",
    "bye": "See ya later ğŸ‘‹",
    "do you play game?": "yep, type 'back' to go to main menu ğŸ®"
}

# 3. player stats
stats = {
    "energy": 100,
    "points": 0,
    "mood": "ğŸ˜„"
}

# 4. greet user
def GreetUser():
    print("\nğŸ‘‹ Welcome to the GenZ Life Simulator.")
    print("Hi", user["name"] + "! Let's get started!")

# 5. responses
def GetResponse(user_input):
    if user_input in responses:
        return responses[user_input]
    else:
        return "hmm I didnâ€™t quite get that ğŸ˜…"

# 6. status & stats
def ShowStatus():
    print("\nğŸ“Š Your current stats:")
    print("âš¡ Energy:", stats["energy"])
    print("ğŸ… Points:", stats["points"])
    print("ğŸ˜„ Mood:", stats["mood"])

def UpdateStats(activity):
    if activity == "study":
        stats["energy"] -= 20
        stats["points"] += 10
        stats["mood"] = "ğŸ“˜ Focused"
        print("You studied hard! ğŸ“š")

    elif activity == "eat":
        stats["energy"] += 15
        stats["mood"] = "ğŸ” Satisfied"
        print("You ate something tasty!")

    elif activity == "play":
        stats["energy"] -= 10
        stats["points"] += 5
        stats["mood"] = "ğŸ® Happy"
        print("You played a game ")

    elif activity == "sleep":
        stats["energy"] = 100
        stats["mood"] = "ğŸ˜´ Rested"
        print("You took a good nap ğŸ’¤")

    elif activity == "garden":
        MaintainGarden()

    else:
        print("Not an option bro ğŸ˜…")


# 7. NEW FEATURE â€” Maintain Garden
def MaintainGarden():
    print("\nğŸŒ± Welcome to your Garden!")
    print("You can plant vegetables (potato/tomato/green chilli). Type 'flower' to move on to flowers.")

    planted_veggies = []

    # vegetable loop
    while True:
        veg = input("\nWhat vegetable would you like to plant? (potato/tomato/green chilli) or 'flower': ").strip()

        if veg == "flower":
            break

        if veg not in ["potato", "tomato", "green chilli"]:
            print("I only have potato, tomato, or green chilli seeds â€” try one of those ğŸ˜…")
            continue

        print(f"\nNow planting {veg.upper()} ğŸŒ¿")

        soil = input("Organic fertilizer or normal soil? (organic/normal): ").strip().lower()
        if soil == "organic":
            print("ğŸŒ¿ Organic fertilizer chosen â€” nutritious soil!")
        else:
            print("Normal soil chosen â€” classic choice!")

        print("\nâ¡ Step 1: Prepare the spot. Type 'done' when finished:")
        input()

        print("â¡ Step 2: Place the seed:")
        input()

        print("â¡ Step 3: Cover lightly with soil:")
        input()

        while True:
            w = input("â¡ Step 4: Water amount in ml: ").strip()
            try:
                water = int(w)
                break
            except:
                print("Enter a number (e.g., 100).")

        if water < 50:
            print("ğŸ’§ Too little water.")
        elif water > 200:
            print("Too much water!")
        else:
            print("ğŸŒŠ Perfect water amount.")

        print("â¡ Step 5: Sunlight or cover:")
        input()

        planted_veggies.append(veg)
        print(f"ğŸ‰ You planted {veg} successfully!")

        again = input("Plant another vegetable? (y/n): ").lower()
        if again != "y":
            break

    print("\nNice job! Vegetables planted:", ", ".join(planted_veggies) if planted_veggies else "none")

    # Flowers
    print("\nNow letâ€™s plant some flowers ğŸŒ¸ğŸŒ¹ğŸŒ»")
    planted_flowers = []

    while True:
        flower = input("\nFlower (rose/tulip/sunflower) or 'n' to stop: ").lower()
        if flower == "n":
            break

        if flower not in ["rose", "tulip", "sunflower"]:
            print("Choose rose, tulip, or sunflower ğŸ˜…")
            continue

        print(f"\nNow planting {flower.upper()} ğŸŒ·")

        print("â¡ Step 1: Prepare pot/bed:")
        input()

        print("â¡ Step 2: Place seed/bulb:")
        input()

        while True:
            w = input("â¡ Step 3: Water amount in ml: ")
            try:
                water = int(w)
                break
            except:
                print("Enter a number (e.g., 100).")

        if water < 50:
            print("ğŸ’§ Too little water.")
        elif water > 200:
            print("Too much water!")
        else:
            print("ğŸŒŠ Good amount!")

        print("â¡ Step 4: Sunlight/shade:")
        input()

        ff = input("â¡ Step 5: Add fertilizer? (y/n): ").lower()
        if ff == "y":
            print("ğŸŒ± Fertilizer added!")
        else:
            print("No fertilizer added.")

        planted_flowers.append(flower)
        print(f"ğŸ‰ You planted a {flower}!")

        again = input("Plant another flower? (y/n): ").lower()
        if again != "y":
            break

    print("\nğŸŒ¼ Garden session complete!")
    print("Vegetables:", ", ".join(planted_veggies) if planted_veggies else "none")
    print("Flowers:", ", ".join(planted_flowers) if planted_flowers else "none")

    stats["points"] += 20
    stats["mood"] = "ğŸŒ» Peaceful"
    print("You earned +20 points and feel relaxed ğŸŒ¸")


# 8. chat
def StartChat():
    print("\nğŸ’¬ Chat mode. (Type 'back' to return)")
    while True:
        user_input = input("You: ").lower()

        if user_input == "back":
            print("Returning to main menu...")
            break

        print("Bot:", GetResponse(user_input))

    print("\nPress 'm' for Main Menu or anything else to exit.")
    go_back = input("Your choice: ").lower()
    if go_back == 'm':
        MainMenu()
    else:
        print("Alright, see ya next time ğŸ‘‹")


# 9. game mode
def StartGame():
    print("\nğŸ® Game mode. (Type 'quit' to go back.)")

    while True:
        ShowStatus()

        print("\nChoose activity:")
        print("1. study ğŸ“˜")
        print("2. eat ğŸ”")
        print("3. play ğŸ®")
        print("4. sleep ğŸ˜´")
        print("5. maintain garden ğŸŒ±")
        print("6. quit")

        choice = input("Enter choice: ").lower()

        if choice in ["quit", "6"]:
            print("Back to main menu...")
            break

        elif choice in ["1", "study"]:
            UpdateStats("study")

        elif choice in ["2", "eat"]:
            UpdateStats("eat")

        elif choice in ["3", "play"]:
            UpdateStats("play")

        elif choice in ["4", "sleep"]:
            UpdateStats("sleep")

        elif choice in ["5", "maintain garden"]:
            UpdateStats("garden")

        else:
            print("try again ğŸ˜…")


# 10. main menu
def MainMenu():
    GreetUser()

    while True:
        print("\nğŸ  Main Menu:")
        print("1. Chat")
        print("2. Play Game")
        print("3. Exit")

        choice = input("Choose: ").lower()

        if choice in ["1", "chat"]:
            StartChat()

        elif choice in ["2", "play"]:
            StartGame()

        elif choice in ["3", "exit"]:
            print("Bye bye ğŸ‘‹")
            break

        else:
            print("Invalid input ğŸ˜…")


# start program
MainMenu()
```
